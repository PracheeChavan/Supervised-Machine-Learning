```python
from sklearn.model_selection import train_test_split
X = df[['Open','High','Low','Close','Adj Close','Volume']]
y = X.pop("Close")

# data splitting
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=123)
```

```python
X_train.head(5)
```

### Comparison of different models

#### Linear Regression

```python
import numpy as np
from sklearn.linear_model import LinearRegression
linear= LinearRegression()
linear.fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score
linear_pred=linear.predict(X_test)
performances = pd.DataFrame(columns=["performance"])
performances.loc["Linear_regression"] = r2_score(y_test, linear_pred)*100
performances
```

#### Decision tree

```python
# Decision tree
from sklearn.tree import DecisionTreeRegressor
tree = DecisionTreeRegressor() # initialize
tree.fit(X_train, y_train)
```

```python
tree_pred = tree.predict(X_test) # Decision tree
performances.loc["Decision Tree"] = r2_score(y_test, tree_pred)*100
performances
```

#### K Nearest neighbors

```python
# K Nearest neighbors
from sklearn.neighbors import KNeighborsRegressor
neigh = KNeighborsRegressor(n_neighbors=1) # initialize with parameters
neigh.fit(X_train, y_train)
```

```python
knn_pred = neigh.predict(X = X_test)
performances.loc["K-Nearest Neighbour"] = r2_score(y_test, knn_pred)*100
performances
```

#### RandomForest regressor

```python
# Decision tree
from sklearn.ensemble import RandomForestRegressor
Rforest = RandomForestRegressor() # initialize
Rforest.fit(X_train, y_train)
```

```python
forest_pred = Rforest.predict(X_test) # Decision tree
performances.loc["Random Forest"] = r2_score(y_test, forest_pred)*100
performances
```

#### Support Vector Regressor

```python
from sklearn import svm
SVC= svm.SVR()
SVC.fit(X_train, y_train)
```

```python
svc_pred = SVC.predict(X_test) # Decision tree
performances.loc["SVC Regressor"] = r2_score(y_test, svc_pred)*100
performances
```
